```python
from transformers import pipeline

# Your category labels
categories = [
    "housing", "utilities", "groceries", "transportation", "health", "debt",
    "dining out", "shopping", "subscriptions", "entertainment", "fees and charges", "education"
]

# Initialize zero-shot classification pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")  # BERT-compatible model

# Classify each transaction
def classify_transaction(text):
    result = classifier(text, categories, multi_label=False)
    return result["labels"][0]  # top predicted label

# Combine description and sub_description
df["full_description"] = df["description"].fillna('') + " " + df["sub_description"].fillna('')

# Predict category
df["predicted_category"] = df["full_description"].apply(classify_transaction)

# Save result or inspect
df[["full_description", "predicted_category"]].head()
```
